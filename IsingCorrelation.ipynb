```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(0,20)
```

```python
y3 = pd.read_csv('IsingCorrelation.txt',sep=',',header=None)
y3 = y3.drop(y3.columns[20],axis=1)
y3 = y3.to_numpy().T
y3data = np.zeros(20)
for i in range(20):
    y3data[i] = y3[i][0]
```

```python
# Using lr = 0.1, epochs = 40, J = 70, L = 40, N = 1000
yRBM = pd.read_csv('RBMCorrelation.txt',sep=',',header=None)
yRBM = yRBM.drop(yRBM.columns[20],axis=1)
yRBM = yRBM.to_numpy().T
yRBMdata = np.zeros(20)
for i in range(20):
    yRBMdata[i] = yRBM[i][0]
```

```python
plt.axhline(y=0,color='black',linestyle='dashed',alpha=0.75)
plt.plot(x,y3,alpha=1,color='red',label='MCMC Ising Chain',linewidth=0.75)
plt.scatter(x,y3,s=35,alpha=1,facecolors='none',edgecolors='red')
plt.plot(x,yRBM,alpha=1,color='blue',label='RBM Network',linewidth=0.75)
plt.scatter(x,yRBM,s=35,alpha=1,facecolors='none',edgecolors='blue',marker='v')
plt.xticks([0,5,10,15,20])
plt.ylim(-0.4,1.1)
plt.xlabel(r'$x$',size=13)
plt.ylabel(r'$C(x)= \langle \sigma_i \sigma_{i+x} \rangle$',size=13)
plt.text(12, 0.5, r'$J_1 = 1 \;\;\;\;\;J_2 = 1/2$',fontsize=13)
plt.title('ANNNI Correlation')
plt.legend(framealpha=0,fontsize=12)
plt.savefig('ANNNI_Correlation.png',bbox_inches='tight', dpi = 1000)
plt.show()
```
